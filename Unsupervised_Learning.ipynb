```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = {
    'Customer':["Rana", "Hamza","Owais", "Claudio", "Miryam", "Khalid"],
    'Age':[25, 30, 35, 40, 45, 50],
    'Spending': [100,200,400,250,140,350],
    }

df = pd.DataFrame(data)
```

```python
X= df[['Age','Spending']]
model = KMeans(n_clusters=2, random_state=42, n_init=10)
df['Group'] =model.fit_predict(X) #fit for training and predict for prediction by trained model
plt.figure(figsize=(6,5)) #6= width 5=height
for group in df['Group'].unique():
  group_data = df[df['Group']==group]
  plt.scatter(group_data['Age'], group_data['Spending'], label=f'Group {group}')

plt.xlabel('Age')
plt.ylabel('Spending')
plt.title('Customer Segmentation')
plt.legend()
plt.grid(True)
plt.show()
print(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler #converts values into z-score so every feature gets equal importance, z-score = (X - Mean)/Standard Deviation
from sklearn.decomposition import PCA

data = {
    'Age':[25, 30, 35, 40, 45, 50],
    'Income':[50000, 60000, 70000, 80000, 90000, 100000],
    'Spending': [100,200,400,250,140,350],
    'Savings': [5000, 10000, 15000, 20000, 25000, 30000]
    }

df = pd.DataFrame(data)

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

pca = PCA(n_components=2) #we are telling to summarize 4 columns into 2
pca_result = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(pca_result, columns=['PC1', 'PC2'])

explained_variance = pca.explained_variance_ratio_ #it tells how much percentage of total information is captured by each component PCA1, PCA2
print("Variance captured by PCA Component:", explained_variance)
print(np.round(explained_variance * 100, 2))

plt.figure(figsize=(8,6))
plt.scatter(pca_df['PC1'], pca_df['PC2'])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Analysis')
plt.grid(True)
plt.show()
print(pca_df)
```
